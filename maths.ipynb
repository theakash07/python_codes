```python
def f(x):
  return 2 * x + 1
x_values = [0, 1, 2, 3]
for x in x_values:
  y = f(x)
  print(y)
```

```python
from sympy import *
x = symbols('x')
f = 2*x + 1
plot(f)
```

```python
from sympy import *
x = symbols('x')
f = x**2 + 10*x + 25
plot(f)
```

```python
from sympy import *
from sympy.plotting import plot3d
x, y, = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

```python
summation = sum(3*i for i in range(1,6))
print(summation)
```

```python
x = [1, 4, 6, 2]
n = len(x)
summation = sum(10*x[i] for i in range(0,n))
print(summation)
```
